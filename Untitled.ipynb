```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/DELL/OneDrive/Documents/dataset.csv")

plt.figure(figsize=(14, 6))

numeric_features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
for i, feature in enumerate(numeric_features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x=df[feature], color="skyblue")
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()

def detect_outliers_iqr(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    return outliers

df_clean = df.dropna(subset=['species'])

outliers_cleaned = {}
for feature in numeric_features:
    outliers_cleaned[feature] = detect_outliers_iqr(df_clean, feature)

for feature, outlier_data in outliers_cleaned.items():
    print(f"Outliers for {feature} (after cleaning):")
    print(outlier_data[['species', feature]])  
    print("\n")

outliers_body_mass_g = df[df['species'].isna() & df['body_mass_g'].notna()]
print(f"Outliers in body_mass_g with NaN species (possible data errors):")
print(outliers_body_mass_g[['species', 'body_mass_g']])
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import seaborn as sns
import matplotlib.pyplot as plt

df_numeric = df.select_dtypes(include=['float64', 'int64'])  

df_numeric = df_numeric.dropna()

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)

pca = PCA(n_components=2)  
pca_result = pca.fit_transform(df_scaled)

pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['species'] = df['species']  

plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='species', data=pca_df, palette='Set2', s=100, alpha=0.7)
plt.title('PCA of Penguin Data (2 Components)', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
plt.legend(title='Species')
plt.show()
print("Explained variance ratio by principal components:")
print(pca.explained_variance_ratio_)

print("\nCumulative explained variance:")
print(pca.explained_variance_ratio_.cumsum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_filtered = df.dropna(subset=['bill_length_mm', 'bill_depth_mm'])

sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df_filtered, hue='species', palette='Set2')
plt.title('Scatter Plot of Bill Length vs Bill Depth', fontsize=16)
plt.xlabel('Bill Length (mm)', fontsize=12)
plt.ylabel('Bill Depth (mm)', fontsize=12)
plt.legend(title='Species')
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df_filtered, palette='Set2')
plt.title('Box Plot of Bill Depth vs Bill Length by Species', fontsize=16)
plt.xlabel('Bill Length (mm)', fontsize=12)
plt.ylabel('Bill Depth (mm)', fontsize=12)
plt.legend(title='Species')
plt.show()


plt.figure(figsize=(8, 6))
sns.violinplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df_filtered, palette='Set2', split=True)
plt.title('Violin Plot of Bill Depth vs Bill Length by Species', fontsize=16)
plt.xlabel('Bill Length (mm)', fontsize=12)
plt.ylabel('Bill Depth (mm)', fontsize=12)
plt.legend(title='Species')
plt.show()


sns.pairplot(df_filtered, vars=['bill_length_mm', 'bill_depth_mm'], hue='species', palette='Set2')
plt.suptitle('Pair Plot of Bill Length vs Bill Depth by Species', fontsize=16)
plt.show()
sns.jointplot(x='bill_length_mm', y='bill_depth_mm', data=df_filtered, kind='scatter', hue='species', palette='Set2')
plt.suptitle('Joint Plot of Bill Length vs Bill Depth (Regression)', fontsize=16)
plt.show()

plt.figure(figsize=(8, 6))
plt.hexbin(df_filtered['bill_length_mm'], df_filtered['bill_depth_mm'], gridsize=30, cmap='Blues')
plt.colorbar(label='Frequency')
plt.title('Hexbin Plot of Bill Length vs Bill Depth', fontsize=16)
plt.xlabel('Bill Length (mm)', fontsize=12)
plt.ylabel('Bill Depth (mm)', fontsize=12)
plt.show()


correlation_matrix = df_filtered[['bill_length_mm', 'bill_depth_mm']].corr()
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Bill Length vs Bill Depth', fontsize=16)
plt.show()
```
